```python
"""Question 2: Pattern & Trend Analysis ● Define at least three features of interest (e.g., age group, gender ratio, urban vs. rural
   clusters, seasonal trends if applicable).
   Ans:- 1) Age Group - i) It can be discovered from data that abnormal results (A2, A3) are more common in older age groups (50+), while
                           normal results (A1) are dominant in younger age groups.
                        ii) This would suggest age is a significant risk factor influencing test outcomes.
         2) Gender Ratio - i) Analysis might show that males constitute a higher proportion of tests overall, but females have a slightly
                              higher rate of certain abnormal results e.g. A2 result.
                           ii) Alternatively, results might reveal gender balance differs by state — e.g., in some regions more women get
                               tested, while in others men dominate.
         3) Urban vs. Rural clusters - i) You could find that some states like Punjab, West Bengal, Karnataka, etc. have higher counts
                                          of abnormal results A2 & A3 compared to others.
                                       ii) Certain cities may emerge as hotspots where abnormal results are concentrated — possibly
                                           linked to local health conditions, environmental factors, or testing coverage. """
```

```python
#Defining abnormal results
abnormal_results = ['A2', 'A3']

#Calculating abnormalilty per state
state_result = lf.groupby('orgState')['result'].apply(
    lambda x: (x.isin(abnormal_results).sum() / len(x)) * 100
)

#Classifying risk profile based on abnormal percentage
def classify_risk(pct):
    if pct >= 50:
        return "High Risk"
    elif pct >= 20:
        return "Medium Risk"
    else:
        return "Low Risk"

#Apply classification
state_risk = state_result.apply(classify_risk)

#Combining into a DataFrame
state_risk_df = pd.DataFrame({
    'abnormal_pct': state_result,
    'risk_level': state_risk
})

#Showing top 10 cities by abnormality
top_states = state_risk_df.sort_values(by='abnormal_pct', ascending=False).head(10)

#Bar chart
plt.figure(figsize=(12,6))
bars = plt.bar(top_states.index, top_states['abnormal_pct'], color='skyblue')

#Color bars by risk level
for i, risk in enumerate(top_states['risk_level']):
    if risk == "High Risk":
        bars[i].set_color("red")
    elif risk == "Medium Risk":
        bars[i].set_color("orange")
    else:
        bars[i].set_color("green")

plt.title("Top 10 States by Abnormal Result Prevalence (A2 & A3)")
plt.xlabel("State")
plt.ylabel("Abnormal Results (%)")
plt.xticks(rotation=90)
plt.show()
```

```python
#Defining abnormal results
abnormal_results = ['A2', 'A3']

#Calculating abnormalilty per city
city_result = lf.groupby('orgCity')['result'].apply(
    lambda x: (x.isin(abnormal_results).sum() / len(x)) * 100
)

#Classifying risk profile based on abnormal percentage
def classify_risk(pct):
    if pct >= 50:
        return "High Risk"
    elif pct >= 20:
        return "Medium Risk"
    else:
        return "Low Risk"

city_risk = city_result.apply(classify_risk)

#Combining into a DataFrame
city_risk_df = pd.DataFrame({
    'abnormal_pct': city_result,
    'risk_level': city_risk
})

#Showing top 10 cities by abnormality
top_cities = city_risk_df.sort_values(by='abnormal_pct', ascending=False).head(10)

#Bar chart
plt.figure(figsize=(12,6))
bars = plt.bar(top_cities.index, top_cities['abnormal_pct'], color='skyblue')

#Color bars by risk level
for i, risk in enumerate(top_cities['risk_level']):
    if risk == "High Risk":
        bars[i].set_color("red")
    elif risk == "Medium Risk":
        bars[i].set_color("orange")
    else:
        bars[i].set_color("green")

plt.title("Top 10 Cities by Abnormal Result Prevalence (A2 & A3)")
plt.xlabel("City")
plt.ylabel("Abnormal Results (%)")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
#Doing analysis by combining Internal and external Data i)
import seaborn as sns

#Cleaning column names
lf.columns = lf.columns.str.strip().str.upper()
lf2.columns = lf2.columns.str.strip().str.upper()

#Standardizing age categories
lf.rename(columns={"AGE": "AGE_GROUP"}, inplace=True)
lf2.rename(columns={"AGE": "AGE_GROUP"}, inplace=True)

#Standardizing gender
lf["GENDER"] = lf["GENDER"].str.strip().str.upper()
lf2["GENDER"] = lf2["GENDER"].str.strip().str.upper()

#Summarizing abnormality by age & gender (uACR data)
abnormal_results = ["A2", "A3"]
med_summary = (lf.assign(ABNORMAL=lf["RESULT"].isin(abnormal_results).astype(int)).groupby(["AGE_GROUP", "GENDER"])["ABNORMAL"].mean()
               .reset_index().rename(columns={"ABNORMAL": "ABNORMAL_RATE"}))

#Convert AGE_GROUP in med_summary to string to match life_summary
med_summary['AGE_GROUP'] = med_summary['AGE_GROUP'].astype(str)


#Convert columns to numeric, coercing errors to NaN
cols_to_numeric = ["FRUITS_VEGGIES", "DAILY_STEPS", "SLEEP_HOURS", "DAILY_STRESS", "WORK_LIFE_BALANCE_SCORE", "WEEKLY_MEDITATION"]
for col in cols_to_numeric:
    lf2[col] = pd.to_numeric(lf2[col], errors='coerce')

life_summary = (lf2.groupby(["AGE_GROUP", "GENDER"])[cols_to_numeric].mean().reset_index())

#Merging the two summaries
combined = pd.merge(med_summary, life_summary, on=["AGE_GROUP", "GENDER"], how="inner")

print(combined.head())
```

```python
#ii) Handling potential NaN or other cases
def map_age_to_range(age):
    if age < 20:
        return 'Less than 20'
    elif 21 <= age <= 35:
        return '21 to 35'
    elif 36 <= age <= 50:
        return '36 to 50'
    elif age >= 51:
        return '51 or more'
    else:
        return None

med_summary['AGE_GROUP_RANGE'] = med_summary['AGE_GROUP'].astype(float).apply(map_age_to_range)

#Now merging using AGE_GROUP_RANGE from med_summary and AGE_GROUP from life_summary
combined = pd.merge(med_summary, life_summary, left_on=['AGE_GROUP_RANGE', 'GENDER'], right_on=['AGE_GROUP', 'GENDER'], how='inner')

print(combined.head())
```
